# Outliers

```python
df.columns
```

```python
#renaming columns
df.columns = ['Площа_заг_м2', 'Колич_сотрудников', 'Тип', 'Кол_чеков', 'Ср_колич_ед_в_чеке', 'Населення', 'Сер_дохід_насел', 'Магаз_моб_звязку', 'Прод_ринок', 'Реч_ринок', 'Транзитна_зупинка_ГТ', 'Зупинка_ГТ']
```

```python
dfm = df
# Outliers
Площа_заг = []
for Площа_заг_м2 in df['Площа_заг_м2']:
    if Площа_заг_м2 > (df['Площа_заг_м2'].mean() + 3*df['Площа_заг_м2'].std()):
        Площа_заг_м2 = (dfm['Площа_заг_м2'].mean() + 3*df['Площа_заг_м2'].std())
    elif Площа_заг_м2 < (df['Площа_заг_м2'].mean() - 3*df['Площа_заг_м2'].std()):
        Площа_заг_м2 = (dfm['Площа_заг_м2'].mean() - 3*df['Площа_заг_м2'].std())
    Площа_заг.append(Площа_заг_м2)
df['Площа_заг_м2'] = Площа_заг
```
